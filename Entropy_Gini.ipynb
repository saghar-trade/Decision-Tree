```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gini(p):
    return p * (1 - p) + (1 - p) * (1 - (1 - p))
```

```python
def entropy(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)
```

```python
def error(p):
    return 1 - np.max([p, 1 - p])
```

```python
x = np.arange(0.001, 1, 0.001)
```

```python
ent = [entropy(p) if p !=0 else None  for p in x]
gin = [gini(p) for p in x]
err = [error(p) for p in x]
```

```python
plt.figure(figsize=(10,6))
ax = plt.subplot(111)

for y, label, ls, color in zip(
    [ent, gin, err],
    ['Entropy', 'Gini Impurity', 'Misclassification Error'],
    ['-', '--', '-.'],
    ['b', 'g', 'r']
):
    ax.plot(x, y, label=label, linestyle=ls, linewidth=2, color=color)

ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')

ax.legend(loc='best')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.title("Comparison of Impurity Metrics")
plt.grid(True)
plt.show()
```
